```python
 def is_valid(s: str) -> bool:
    open_parentheses = {'(': 0, '{': 0, '[': 0}
    close_parentheses = {')': 0, '}': 0, ']': 0}
    for char in s:
        if char in open_parentheses:
            open_parentheses[char] += 1
        elif char in close_parentheses:
            # This line was not indented correctly
            close_parentheses[char] += 1
            if close_parentheses[char] > open_parentheses[char]:  # More closing than opening
                return False

    # Check if all parentheses are balanced at the end
    for char in open_parentheses:
        if open_parentheses[char] != close_parentheses[char.replace('(', ')').replace('{', '}').replace('[', ']')]:
            return False
    return True  # All parentheses are balanced
    return open_parentheses == close_parentheses
```
