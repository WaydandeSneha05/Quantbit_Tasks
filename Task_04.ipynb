```python
def longest_consecutive(nums: list) -> int:
    if not nums:
        return 0

    nums.sort()

    max_len = 1
    current_len = 1


    for i in range(1, len(nums)):
        if nums[i] == nums[i - 1]:
            continue
        elif nums[i] == nums[i - 1] + 1:
            current_len += 1
        else:

            max_len = max(max_len, current_len)
            current_len = 1


    return max(max_len, current_len)
```

```python
nums = [100, 4, 200, 1, 3, 2]
print(longest_consecutive(nums))
```
